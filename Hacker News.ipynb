# Hacker News Posts


 Hacker news is a site started by the startup incubator Y Combinator, where user-submitted stories (known as 'posts') are voted and commented upon, similar to reddit. Hacker News is extremely popular in technology and startup circles, and posts that make it to the top of Hacker News' listings can get hundreds of thousands of visitors as a result.

 The dataset for the following project is [here](https://www.kaggle.com/hacker-news/hacker-news-posts) where it has approximately 300,000 rows.

 We are specifically intreste in posts whose titles begin with either Ask HN or show HN.

```python
# Displaying first five rows

opened_file = open('HN_posts_year_to_Sep_26_2016.csv',encoding='utf8')
from csv import reader
read_file = reader(opened_file)
hn = list(read_file)
hn_header = hn[0]
hn = hn[1:]
print(hn[:5])
```

* Here we are going to separete the header row from rest of the dataset and display them below
* We are going to analyse total number of rows and columns

```python
def explore_dataset(dataset,start, end, rows_and_columns = False):
    data_slice = dataset[start:end]
    for row in data_slice:
        print(row)
        print('\n') # It adds spacing between each row
        
    if rows_and_columns:
        print('Number of columns :', len(dataset[0]))
        print('Number of rows :', len(dataset))

print('\n')
print(hn_header)
print('\n')
explore_dataset(hn, 0, 5, rows_and_columns = True)
```

Next we are going to create three empty list called ask_posts, show_posts and other_posts. We are going to iterate over hacker news post and seprate them into three different list, using append function.

We are also going to print first row of ask posts and show posts and their total length.

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    if (title.startswith('Ask HN')):
        ask_posts.append(row)
    elif (title.startswith('Show HN')):
        show_posts.append(row)
    else:
        other_posts.append(row)
      
    
print('First row of ask posts:')
print(ask_posts[0])
print('\n')
print('length of ask posts:',len(ask_posts)); print('\n')

print('First row of show posts:')
print('First row of show posts',show_posts[0])
print('\n')
print('length of show posts:', len(show_posts)); print('\n')

print('length of other posts:',len(other_posts)) 
        
```

### Average of Ask post and Show posts
Now as we have seperated show posts and ask posts we are going to calculate total number of comments in show posts and ask posts, as well as their average.

* first we are going to initialize total_ask_comments ans total_show_comments to zero. 
* And later loop the ask post and calculate total number of ask and show comments
* There after we are going to calculate the average of both show post and ask post by dividing total ask comments and total show comments by length of ask posts and show posts respectively.

```python
total_ask_comments = 0
for row in ask_posts:
    num_comments = int(row[4])
    total_ask_comments += num_comments
    
avg_ask_comments = total_ask_comments/ len(ask_posts)
print('Total ask comments:',total_ask_comments) 
print('Avg ask comments:',avg_ask_comments)
print('\n')
total_show_comments = 0
for row in show_posts:
    num_comments = int(row[4])
    total_show_comments += num_comments
    
avg_show_comments = total_show_comments/ len(show_posts)
print('Total show comments',total_show_comments)
print('Avg show comments',avg_show_comments)
```

 Since ask posts are more likely to receive comments, we'll focus our remaining analysis just on these posts.
Next, We will calculate the amount of ask posts and comments by hour created. We'll use the datetime module to work with the data in the created_at column.

    We will first import datetime module as well as create an empty result list. Iterate over ask_posts and append to result_list. Then create two dictionaries count_by_hours and comment_by_hours and iterate over result_list, if hours isn't a key in count_by_hours then set it equal to 1 or if already in count_by_hours increase by 1 for each iteration.

```python
import datetime as dt
result_list = []
for row in ask_posts:
    created_at = row[6]
    num_comments = int(row[4])
    result_list.append([created_at,num_comments])
    
print(result_list[:4])   

count_by_hours = {}
comments_by_hours = {}

for row in result_list:
    dt_object = dt.datetime.strptime(row[0], "%m/%d/%Y %H:%M")
    dt_hours = dt_object.strftime("%H")
    if dt_hours not in count_by_hours:
        count_by_hours[dt_hours] = 1
        comments_by_hours[dt_hours] = row[1]
    else:
        count_by_hours[dt_hours] += 1
        comments_by_hours[dt_hours] += row[1]
print('\n')         
print("Count by hours :",count_by_hours)
print('\n') 
print("Comments every hour",comments_by_hours)
```

Next, we create an empty avg_by_hours list and calculate the average number of comments per post for posts created during each hour of the day and append it to avg_by_hours list. The result will be a list of lists in which the first element is the hour and the second element is the average number of comments per post. 

```python
avg_by_hours = []
for row in comments_by_hours:
    for row1 in count_by_hours:
        avg_by_hours.append([row, comments_by_hours[row]/count_by_hours[row1]])


print(avg_by_hours)       
```

Although we now have the results we need, this format makes it hard to identify the hours with the highest values. Let's finish by sorting the list of lists and printing the five highest values in a format that's easier to read.
 
 * Create a list that equals avg_by_hour with swapped columns.
 * Use the sorted() function to sort swap_avg_by_hour in descending order. 


```python
swap_avg_by_hours = []
for row in avg_by_hours:
    a = row[0]
    row[0] = row[1]
    row[1] = a
    swap_avg_by_hours.append([row[0], row[1]])
    
print(swap_avg_by_hours)        
```

```python
sorted_swap = sorted(swap_avg_by_hours, reverse = True)
print(sorted_swap)
```

```python
print("Top 5 Hours for Ask Posts Comments")
```

```python
Next, we will Loop through each average and each hour in the first five lists of sorted_swap.
```

```python
import datetime as dt
for row in sorted_swap:
    hours = row[1]
    num_comments = row[0]
    hour_object = dt.datetime.strptime(hours, "%H")
    hour_string = hour_object.strftime("%H")
    
    print("{}: {:.2f} average comments per posts".format(hour_string, num_comments))
```

So, In the end as you can see from the above output, the posts that are mostly created at 3 PM received highest number of comments.
